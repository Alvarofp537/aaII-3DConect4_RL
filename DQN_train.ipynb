```python
import numpy as np
import plotly.graph_objects as go
from typing import List, Dict

def visualize_3d_game(boards: List[np.ndarray], mapping: Dict[int, str]):
    # Colores base por jugador (puedes ajustar)
    COLORS = {
        0: "rgba(0,0,0,0)",  # vacío (transparente)
        1: "red",
        2: "blue",
        3: "green",
        4: "orange",
    }

    frames = []
    max_z, max_x, max_y = boards[0].shape

    for i, grid in enumerate(boards):
        # Para cada frame, construimos las esferas
        data = []
        for z in range(max_z):
            for x in range(max_x):
                for y in range(max_y):
                    player = grid[z, x, y]
                    if player == 0:
                        continue
                    data.append(go.Scatter3d(
                        x=[x], y=[y], z=[z],
                        mode="markers",
                        marker=dict(
                            size=8,
                            color=COLORS.get(player, "gray"),
                            opacity=0.9,
                        ),
                        name=mapping.get(player, f"Player {player}"),
                        showlegend=False
                    ))
        frames.append(go.Frame(data=data, name=str(i)))

    # Figura base
    layout = go.Layout(
        scene=dict(
            xaxis=dict(title="X", range=[0, 6]),  # Ancho fijo de 0 a 6
            yaxis=dict(title="Y", range=[0, 3]),  # Profundidad fija de 0 a 3
            zaxis=dict(title="Z", range=[0, 5]),  # Altura fija de 0 a 5
        ),
        margin=dict(l=0, r=0, b=0, t=30),
        sliders=[{
            "steps": [
                {
                    "method": "animate",
                    "label": f"{i}",
                    "args": [[str(i)], {"frame": {"duration": 0, "redraw": True},
                                        "mode": "immediate"}],
                } for i in range(len(frames))
            ],
            "transition": {"duration": 0},
            "x": 0.1,
            "xanchor": "left",
            "y": 0,
            "yanchor": "top"
        }],
        updatemenus=[{
            "buttons": [
                {
                    "args": [None, {"frame": {"duration": 500, "redraw": True},
                                    "fromcurrent": True, "mode": "immediate"}],
                    "label": "▶ Play",
                    "method": "animate"
                },
                {
                    "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                    "mode": "immediate"}],
                    "label": "⏸ Pause",
                    "method": "animate"
                }
            ],
            "type": "buttons",
            "direction": "left",
            "pad": {"r": 10, "t": 87},
            "showactive": False,
            "x": 0.1,
            "xanchor": "right",
            "y": 0,
            "yanchor": "top"
        }]
    )


    # Añadir la leyenda aparte (primer frame + leyenda única)
    initial_data = []
    for pid, name in mapping.items():
        initial_data.append(go.Scatter3d(
            x=[None], y=[None], z=[None],
            mode="markers",
            marker=dict(size=8, color=COLORS.get(pid, "gray")),
            name=name,
            showlegend=True
        ))

    initial_data += frames[0].data

    fig = go.Figure(data=initial_data, layout=layout, frames=frames)
    fig.update_layout(title="Connect-N 3D Game Replay")

    fig.show()
```
